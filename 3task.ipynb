```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
import statsmodels.formula.api as smf
from pylab import *
%matplotlib inline

df = pd.read_excel('Analyst_Data.xls', sheetname=3)
df = df[['X', 'Y']]
print(df.head(5))
```

## Строим модель исходя из имеющихся данных X и Y

```python
plt.scatter(df['X'], df['Y'], linewidths=0.001, marker=".")
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
plt.show()
```

## Строим регрессионную модель с помощью statsmodels.formula.api

```python
df.plot('X', 'Y', kind='scatter', linewidths=0.6)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)

lm = smf.ols('Y ~ X', data=df).fit()
xmin = df.X.min()
xmax = df.X.max()
X = np.linspace(xmin, xmax, 100)
Y = lm.params[0] + lm.params[1] * X
plt.plot(X, Y, color='red')
```
